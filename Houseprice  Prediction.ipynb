```python
#  Cross Validation
```

```python
df_train.head()
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
from sklearn.metrics import make_scorer, r2_score
```

```python
def test_model(model, x_df_train=x_df_train, y_df_train=y_df_train):
        cv = KFold(n_splits=3 , shuffle=True, random_state=45)
        r2 = make_scorer(r2_score)
        r2_val_score = cross_val_score(model, x_df_train, y_df_train, cv=cv , scoring= r2)
        score = [r2_val_score.mean()]
        return score
```
